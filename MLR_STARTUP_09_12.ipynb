```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
```

```python
# SPLIT THE COLUMNS
X = df.iloc[: , : -1]
y = df.iloc[: ,-1]
len(X),len(y)
```
